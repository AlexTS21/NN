```python
import layer as nt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Aproximacion de f(x)=x^2

```python
#Generar conjunto de puntos
def generar_datos(n):
    # Crear el arreglo patrón del 1 a n (columna)
    patron = np.arange(-n, n + 1).reshape(-1, 1)
    # Crear el arreglo s con los cuadrados como float (columna)
    s = (patron ** 2).astype(float)
    return patron, s

patron, s = generar_datos(20)

#Gerar red y entrenar
nn =  nt.nn([1,20,30,1], ["tanh","relu","identidad"])
nn.train(patron, s, epoch=250, learningRate=0.01, normalizar=True)

#Generar los puntos de la red
g = s.copy()
for i in range(len(s)):
    aux = nn.evaluate(patron[i], normalizar=True)
    g[i] = aux
#print(g)
#Generar graficas
plt.figure(figsize=(8, 5))
plt.plot(patron, s, label='original')
plt.plot(patron, g, label='red')


plt.title('x^2 generado vs real')
plt.xlabel('X')
plt.ylabel('s|y')
plt.grid(True)
plt.legend()
plt.tight_layout()

# Show plot
plt.show()
```

Aproximacion de sen

```python
#Generar conjunto de puntos
def generar_datos(n_puntos):
    # Generar 100 puntos de -2π a 2π
    patron = np.linspace(-np.pi, np.pi, n_puntos).reshape(-1, 1)
    # Calcular el seno
    s = np.sin(patron).astype(float)
    return patron, s

patron, s = generar_datos(100)

#Gerar red y entrenar
nn = nt.nn([1, 16, 8, 1], ["tanh", "tanh", "identidad"])
nn.train(patron, s, epoch=1000, learningRate=0.0001, normalizar=True )

#Generar los puntos de la red
g = s.copy()
for i in range(len(s)):
    aux = nn.evaluate(patron[i], normalizar=True)
    g[i] = aux
print(g)
#Generar graficas
plt.figure(figsize=(8, 5))
plt.plot(patron, s, label='original')
plt.plot(patron, g, label='red')


plt.title('sin generado vs real')
plt.xlabel('X')
plt.ylabel('s|y')
plt.grid(True)
plt.legend()
plt.tight_layout()

# Show plot
plt.show()
```

```python
#Para el csv de diavetes
def cargar_datos_desde_csv(ruta_csv):
    # Leer CSV con pandas
    df = pd.read_csv(ruta_csv)

    # Separar entradas y salidas
    patron =  df.drop(columns=['id', 'class'], errors='ignore').values.astype(float)
    s = df[['class']].values.astype(float)  # Mantener como matriz columna

    return patron, s

patron, s = cargar_datos_desde_csv('ED.csv')


nn = nt.nn([16,8,1], ["relu", "relu", "sigmoide"])
nn.train(patron, s, epoch=50, learningRate=0.01, quiet=False)
```

```python
patron, S = cargar_datos_desde_csv('PD.csv')

acierto = 0
for i in range(len(patron)):
    pred = nn.evaluate(patron[i])
    print("Pred:", pred, "Clase:", round(pred[0]))
    if round(pred[0]) == S[i]:
        acierto+=1

print(f'El modelo acerto: {acierto} presicion {acierto/len(patron)}')

    
```

```python
def cargar_datos_desde_csv_Split(ruta_csv, val=0.8):
    # Leer CSV con pandas
    df = pd.read_csv(ruta_csv)

    # Separar entradas y salidas
    patron =  df.drop(columns=['id', 'class'], errors='ignore').values.astype(float)
    s = df[['class']].values.astype(float)  # Mantener como matriz columna
    trainX = patron[:int(len(patron)*val)]
    trainS = s[:int(len(patron)*val)]
    testX = patron[int(len(patron)*val):]
    testS = s[int(len(patron)*val):]
    return trainX, trainS, testX, testS

trainX, trainS, testX, testS =cargar_datos_desde_csv_Split("PD.csv")
print(trainX.shape)
print(testX.shape)
```

```python
import pandas as pd
import numpy as np

def transformar_csv_criterios(ruta_entrada, ruta_salida):
    df = pd.read_csv(ruta_entrada)

    def transformar_valor(val):
        if isinstance(val, str):
            val_lower = val.strip().lower()
            if val_lower == 'yes':
                return 1
            elif val_lower == 'no':
                return 0
            elif val_lower == 'positive':
                return 1
            elif val_lower == 'negative':
                return 0
            elif val_lower == 'female':
                return 1
            elif val_lower == 'male':
                return 0
            else:
                try:
                    num = float(val)
                    return 1 if num > 50 else 0
                except:
                    return val  # si no se puede transformar, lo deja igual
        elif isinstance(val, (int, float, np.number)):
            return 1 if val > 50 else 0
        else:
            return val  # deja el valor si no entra en ningún criterio

    # Aplicar transformación a todos los elementos
    df_transformado = df.applymap(transformar_valor)

    # Guardar el resultado
    df_transformado.to_csv(ruta_salida, index=False)
    print(f"Archivo transformado guardado en: {ruta_salida}")


transformar_csv_criterios("DiabetesE.csv", "ED.csv")
```
